### `Book Store`
`Get the title of books in the store`

```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
```

```python
page_no=12
base_url.format(page_no)
```

```python
import requests
import bs4
```

```python
result = requests.get(url=base_url.format(1))
```

```python
result.text
```

```python
soup = bs4.BeautifulSoup(markup=result.text, features='lxml')
```

```python
soup
```

```python
products = soup.select(selector='.product_pod')
```

```python
products
```

```python
products[0]
```

```python
soup.select(selector='.product_pod > p')
```

```python
soup.select(selector='.product_pod a')
```

```python
soup.select(selector='.product_pod h3 > a')[0].getText()
```

```python
two_star_titles = []
for n in range(1,51):
  scrape_url = base_url.format(n)
  res = requests.get(scrape_url)

  soup = bs4.BeautifulSoup(markup=res.text, features='lxml')
  books = soup.select(".product_pod")

  for book in books:
    if len(book.select('.star-rating.Two')) != 0:
      book_title = book.select('a')[1]['title'] # important index start with 1 not 0
      print(book_title)
      two_star_titles.append(book_title)
```

```python
two_star_titles
```

```python
one_star_titles = []
two_star_titles = []
three_star_titles = []
four_star_titles = []
five_star_titles = []

for i in range(1,51):
  res = requests.get(url=base_url.format(i))
  # parse 
  soup = bs4.BeautifulSoup(markup=res.text, features='lxml')
  for book in soup.select('.product_pod'):
    # now check for stars and store in respective array list
    if len(book.select('.star-rating.Two')) != 0:
      book_title = book.select('a')[1]['title']
      two_star_titles.append(book_title)
    elif len(book.select('.star-rating.Three')) != 0:
      book_title = book.select('a')[1]['title']
      three_star_titles.append(book_title)
    elif len(book.select('.star-rating.Four')) != 0:
      book_title = book.select('a')[1]['title']
      four_star_titles.append(book_title)
    elif len(book.select('.star-rating.Five')) != 0:
      book_title = book.select('a')[1]['title']
      five_star_titles.append(book_title)
    else:
      book_title = book.select('a')[1]['title']
      one_star_titles.append(book_title)
```

```python
four_star_titles
```

```python
five_star_titles
```

```python
one_star_titles
```

```python
# lets create a list of set having quote & author
base_url = 'http://quotes.toscrape.com/page/{}/'
page = 1
quotes_set = set()
#requests.get(url=base_url.format(page))

while requests.get(url=base_url.format(page)):
  print(base_url.format(page))
  res = requests.get(url=base_url.format(page))
  if 'No quotes found!' in res.text:
    break
  soup = bs4.BeautifulSoup(markup=res.text, features='lxml')

  for quote in soup.select('.quote'):
    quotes_set.add(quote.select('.text')[0].getText())
    quotes_set.add(quote.select('.author')[0].getText())
  page+=1
```
